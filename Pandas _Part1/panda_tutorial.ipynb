```python
# Cleaning Data
df = pd.read_csv("raw_data.csv")
df2 = df.copy()
df2["tax"] = df2["income"].apply(lambda x : "20%" if x >= 60000 else"10%")
df2
```

```python
df2 = df.copy()

new_col_order = [col for col in df2.columns if col != "id" ] + ["id"]
print (new_col_order)
df[new_col_order]
```

```python
df2 = df.copy()

df2 = df2.drop_duplicates()
df2 = df2.fillna(0)

df2= df2.sort_values("income")

df2= df2.reset_index(drop = True)
df2.to_csv("sorted_data.csv")
df2
```

```python
df = pd.DataFrame({
    "country": ["USA", "USA", "India", "India"],
    "year": [2020, 2021, 2020, 2021],
    "sales": [100, 120, 90, 110],
    "profit": [20, 25, 18, 22],
    "tax" : [10,20,30,40]
})

df.melt(
    id_vars = ["country","year"],
    value_vars =["sales","profit","tax"],
    var_name = "matrix",
    value_name = "value"
)
```
